```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import requests
from bs4 import BeautifulSoup
import re
```

# 2. Instantiate Model

```python
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')

model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

# 3. Encode and Calculate Sentiment

```python
tokens = tokenizer.encode('Checked Enish out on a strong recommendation of several Nigerian friends from the local area in London. My first venture into Nigerian cuisine. I came in with recommendations which I bounced off of the gentleman behind the counter of the restaurant. He was pretty helpful, but could have been much more customer-focused relative to the service. I ordered the tilapia with agusi and pounded yam. Very good, spicy the way I like things... but good. I give 3 starts due to the lack of good service. Once I ordered the guy completely ignored me until I was ready to pay... so much so that I didnt even get to order a drink of any kind. If you like Nigerian or are curious... Id recommend checking it out', return_tensors='pt')
```

```python
result = model(tokens)
```

```python
result.logits
```

```python
int(torch.argmax(result.logits))+1
```

# 4. Collect Reviews

```python
r = requests.get('https://www.yelp.com/biz/social-brew-cafe-pyrmont')
soup = BeautifulSoup(r.text, 'html.parser')
regex = re.compile('.*comment.*')
results = soup.find_all('p', {'class':regex})
reviews = [result.text for result in results]
```

# 5. Load Reviews into DataFrame and Score

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.array(reviews), columns=['review'])
```

```python
df['review'].iloc[0]
```

```python
df.head()
```

```python
df.info()
```

```python
def sentiment_score(review):
    tokens = tokenizer.encode(review, return_tensors='pt')
    result = model(tokens)
    return int(torch.argmax(result.logits))+1
```

```python
sentiment_score(df['review'].iloc[1])
```

```python
sentiment_score(df['review'].iloc[2])
```

```python
df['sentiment'] = df['review'].apply(lambda x: sentiment_score(x[:512]))
```
