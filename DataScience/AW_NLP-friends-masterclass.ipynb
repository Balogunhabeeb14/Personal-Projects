```python
import collections
import nltk
from nltk.tokenize import word_tokenize
```

```python
words = nltk.tokenize.word_tokenize(data_friends)
words[0:20]
```

Then, you use collection to compute the frequency for each word:

```python
words_freq = collections.Counter(words)
words_freq.most_common(20)
```

**It seems that we need to apply some cleaning techniques!**

```python
from nltk.corpus import stopwords

#we can check what kind of words are in the list of stopwords by default
stopwords.words('english')
```

```python
words_cleaned = [word for word in words if word not in stopwords.words('english')]
```

```python
words_cleaned[0:50]
```

```python
words_freq = collections.Counter(words_cleaned)
words_freq.most_common(15)
```

# Let's clean up the text a little more

#### Lower case

```python

words_cleaned = [word.lower() for word in words_cleaned]

words_cleaned = [word for word in words_cleaned if word not in stopwords.words('english')]

words_freq = collections.Counter(words_cleaned)
words_freq.most_common(15)
```

#### Take only alpha characters

```python
words_cleaned = [word for word in words_cleaned if word.isalpha()]
words_cleaned[0:15] #show the first 15 words
```

```python
words_freq = collections.Counter(words_cleaned)
words_freq.most_common(15)
```

#### Create our own blacklist

```python
black_list = ["monica", "ross", "chandler", "joey", "rachel", "phoebe"]
words_cleaned = [word for word in words_cleaned if word not in black_list]
words_freq = collections.Counter(words_cleaned)
top_15 = words_freq.most_common(15)
top_15
```

# Plot the frecuency for the first 15 words

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
figure(figsize=(20, 4))
labels = [item[0] for item in top_15]
number = [item[1] for item in top_15]
nbars = len(top_15)
plt.bar(np.arange(nbars), number, tick_label=labels)
plt.show()
```

# Tagging
https://www.guru99.com/pos-tagging-chunking-nltk.html

```python
words_cleaned_tags = nltk.pos_tag(words_cleaned)
words_cleaned_tags[0:20]
```

Warning: thanksgiving!!! is not a verb but because of the 'ing' it is classified as a verb 

### Select only the nouns (singular and plural)
('gang', 'NN')  
word_tag[0] - returns the first element, e.g 'gang'  
word_tag[1] - returns the second element e.g 'NN'  

```python
words_cleaned_NNs = [word_tag[0] for word_tag in words_cleaned_tags if (word_tag[1]=="NNS") or (word_tag[1]=="NN")]
words_cleaned_NNs[0:20]
```

Check the most frequent nouns

```python
words_freq = collections.Counter(words_cleaned_NNs)
top_15 = words_freq.most_common(15)
top_15
```

```python
figure(figsize=(20, 4))
labels = [item[0] for item in top_15]
number = [item[1] for item in top_15]
nbars = len(top_15)
plt.bar(np.arange(nbars), number, tick_label=labels)
plt.show()
```

### Lemmatization and stemming

Stemming and Lemmatization are Text Normalization (or sometimes called Word Normalization) techniques in the field of Natural Language Processing that are used to prepare text, words, and documents for further processing.


Stemming and Lemmatization have been studied, and algorithms have been developed in Computer Science since the 1960's. In this tutorial you will learn about Stemming and Lemmatization in a practical approach covering the background, some famous algorithms, applications of Stemming and Lemmatization, and how to stem and lemmatize words, sentences and documents using the Python nltk package which is the Natural Language Tool Kit package provided by Python for Natural Language Processing tasks.

Stemming and Lemmatization are itself form of NLP and widely used in Text mining. Text Mining is the process of analysis of texts written in natural language and extract high-quality information from text. It involves looking for interesting patterns in the text or to extract data from the text to be inserted into a database. 

```python
words_cleaned[0:20]
```

```python
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
words_cleaned_stemm = [porter_stemmer.stem(word) for word in words_cleaned]
words_cleaned_stemm[0:20]
```

Warining! some of them can be incorrect 

```python
from nltk.stem import LancasterStemmer
lancaster = LancasterStemmer()
words_cleaned_lancaster=[lancaster.stem(word) for word in words_cleaned]
words_cleaned_lancaster[0:20]
```

```python
#give another exemple with context
word_list = ["friend", "friendship", "friends", "friendships","stabil","destabilize","misunderstanding","railroad","moonlight","football"]
print("{0:20}{1:20}{2:20}".format("Word","Porter Stemmer","lancaster Stemmer"))
for word in word_list:
    print("{0:20}{1:20}{2:20}".format(word,porter_stemmer.stem(word),lancaster.stem(word)))
```

The differencw with lemmatizer is that it gets the context, and for verbs sometimes gets the ing as adj..

### N-GRAM

The essential concepts in text mining is n-grams, which are a set of co-occurring or continuous sequence of n items from a sequence of large text or sentence. The item here could be words, letters, and syllables. 1-gram is also called as unigrams are the unique words present in the sentence.

```python
#bigrams, make assosiations of two words that happen together
from nltk import ngrams
bigrams = ngrams(data_friends.split(), 2)

show=0
for bigram in bigrams:
    if(show<10):
        print(bigram)
        show +=1
```

```python
from nltk import ngrams
trigrams = ngrams(data_friends.split(), 3)
show=0
for trigram in trigrams:
    if(show<10):
        print(trigram)
        show +=1
```

## SENTIMENT ANALYSIS

```python
#pip install vaderSentiment
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
## Example: We will use the polarity_scores() method to obtain the polarity indices for the given sentence.
def sentiment_analyzer_scores(sentence):
    score = analyzer.polarity_scores(sentence)
    print("{:-<40} {}".format(sentence, str(score)))
```

```python
sentiment_analyzer_scores("The dog likes to play with the ball")
```

The Positive, Negative and Neutral scores represent the proportion of text that falls in these categories.

The Compound score is a metric that calculates the sum of all the lexicon ratings which have been normalized between -1(most extreme negative) and +1 (most extreme positive).

```python
sentiment_analyzer_scores("I love it!!! ")
```

```python
sentiment_analyzer_scores("That is sooo sad")
```

```python
print(sentiment_analyzer_scores('I am 😄 today'))
print(sentiment_analyzer_scores('😊'))
print(sentiment_analyzer_scores('I am 😥 for the remote classes'))
print(sentiment_analyzer_scores('I am ☹️ for the remote classes'))
print(sentiment_analyzer_scores('Oh no! This is the worst thing I ever heard ☹️☹️☹️'))
```
