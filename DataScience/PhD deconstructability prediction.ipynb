### Process the data 

As observed from the data, most features with object types are categorical and needs to be preprocessed before using machine learning on them.

what is categorical feature? Categorical data is non-numeric and often can be characterized into categories or groups. A simple example is the option at the EOL;renovation, deconstrcution or demolition. Another example is the  interval-type data. Like 2-5days, 6-10days etc., and so on. The intervals are categorical. 

- why Encode categorical data?
- We encode categorical data numerically because math is generally done using numbers. A big part of natural language processing is converting text to numbers. Just like that, our algorithms cannot run and process data if that data is not numerical. 

- Therefore, we need to have tools at our disposal to transform colors like red, yellow, and blue into numbers like 1, 2, and 3 for all the backend math to take place. Now that we know what categorical data looks like and have seen some examples, we will examine three common methods to turn our categorical data into numeric data.

The convertion of the categorical data to numeric can be done using 
- one hot, label and target encoder.

- we will be using label encoder
